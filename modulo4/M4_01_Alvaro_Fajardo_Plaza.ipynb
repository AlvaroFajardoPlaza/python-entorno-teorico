```python
class Vehiculo():
    """Clase Vehículo
    
    Esta clase incluye dos argumentos: marca y modelo, de tipo string ambos. También incluye el método repostar() que imprime por pantalla un texto que dice: 'Este vehiculo tiene que repostar gasolina.'
    """
    # Constructor
    def __init__(self, marca, modelo):
        self.__marca = marca
        self.__modelo = modelo
        print("Vehículo creado.")
    
    # Getters y Setters
    @property
    def marca(self):
        return self.__marca
    
    @marca.setter
    def marca(self, newValue):
        self.__marca = newValue
    
    @property
    def modelo(self):
        return self.__modelo
    
    @modelo.setter
    def modelo(self, newValue):
        self.__modelo = newValue

    # Métodos
    def repostar(self):
        print("\tEste vehiculo tiene que repostar gasolina.")
    
    def __str__(self):
        return "La marca del vehículo es: {}. Su modelo es: {}.".format(self.marca, self.modelo)

class VElectrico():
    """Clase Vehículo Eléctrico
    
    Esta clase incluye tres argumentos: marca, modelo, de tipo string ambos y autonomia, de tipo integer. También incluye el método repostar() que imprime por pantalla un texto que dice: 'Este vehiculo tiene que recargar electricidad.'
    """
    # Constructor
    def __init__(self, marca, modelo, autonomia):
        self.__marca = marca
        self.__modelo = modelo
        self.__autonomia = autonomia
        print("Vehículo Eléctrico creado.")
    
    # Getters y Setters
    @property
    def marca(self):
        return self.__marca
    
    @marca.setter
    def marca(self, newValue):
        self.__marca = newValue
    
    @property
    def modelo(self):
        return self.__modelo
    
    @modelo.setter
    def modelo(self, newValue):
        self.__modelo = newValue
    
    @property
    def autonomia(self):
        return self.__autonomia
    
    @autonomia.setter
    def autonomia(self, newValue):
        self.__autonomia = newValue

    # Métodos
    def repostar(self):
        print("\tEste vehiculo tiene que recargar electricidad.")
    
    def __str__(self):
        return "La marca del vehículo es: {}. Su modelo es: {}, con una autonomia de {}".format(self.marca, self.modelo, self.autonomia)
    
# Creamos la clase Bicicleta eléctrica, priorizando como clase padre la de VElectrico():
class BicicletaElectrica(VElectrico, Vehiculo):
    """
    Clase Bicileta Eléctrica
    
    Esta clase experimenta herencia múltiple de las clases VElectrico y Vehiculo, priorizando la primera. Es por ello que contará con tres atributos: marca, modelo y autonomía.
    """
        # Constructor
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)
        print("Bicicleta eléctrica creada.")
    
    # Métodos
    def __str__(self):
        return "\nDatos de la bicleta eléctrica:\n-> Marca:\t{}\n-> Modelo:\t{}\n-> Autonomía:\t{}kms".format(self.marca, self.modelo, self.autonomia)

# Creamos la clase Quad, priorizando como clase padre la de Vehículo():
class Quad(Vehiculo, VElectrico):
    """
    Clase Quad
    
    Esta clase experimenta herencia múltiple de las clases Vehiculo y VElectrico, priorizando la primera. También contará con tres atributos: marca, modelo y autonomía
    """
    # Constructor
    def __init__(self, marca, modelo, autonomia):
        Vehiculo.__init__(self, marca, modelo) # Para manejar sin confusiones la herencia múltiple, nos referimos a las superclases 'padre' por su nombre.
        VElectrico.__init__(self, marca, modelo, autonomia) # No olvidamos pasar todos los argumentos correspondientes.
        print("Quad creado.")
    
    # Métodos
    def __str__(self):
        return "\nDatos del Quad:\n-> Marca:\t{}\n-> Modelo:\t{}\n-> Autonomía:\t{}kms".format(self.marca, self.modelo, self.autonomia)


# Creamos dos objetos y preguntamos por el método repostar para cada uno de ellos.
biciElec1 = BicicletaElectrica("Speed", "Urban", 250)
quad1 = Quad(marca="Kawasaki", modelo="especial", autonomia=80)

print(biciElec1)
biciElec1.repostar()

print(quad1)
quad1.repostar()




```

```python
# La superclase Personal_Universitario tendrá un solo atributo que será un diccionario >>> Objetos dentro de objetos >>> encapsulación
class Personal_Universitario():
    """Superclase Personal Universitario
    
    Esta super clase unicamente recibe un atributo 'datos', que será un diccionario.
    Dentro de este diccionario podremos encontrer pares de key-value, cuyas keys serán:
    - id: de tipo integer. Código numérico de dos dígitos para cada objeto.
    - nombre: de tipo string. Para definir el nombre de la persona.
    - email: de tipo string. Un email al azar que cumpla con el patrón para considerarse como tal.
    """
    # Constructor
    def __init__(self, id, nombre, email):
        self.__datos = {
            'id': id,
            'nombre': nombre,
            'email': email
        }
        print("Personal Universitario creado")
    
    @property
    def datos(self):
        return self.__datos

    @datos.setter
    def datos(self, newValues): # NewValues tendrá que ser obligatoriamente un diccionario con la misma estructura que acepta
        self.__datos = newValues

    def __str__(self):
        return "\nPersonal Universitario creado con los siguientes datos:\n\t-> ID: {}\n\t-> Nombre: {}\n\t-> Email: {}".format(self.__datos['id'], 
                                                                                                                                self.__datos['nombre'],
                                                                                                                                self.__datos['email'])

class Oficina(Personal_Universitario):
    """Subclase Oficina
    
    Esta subclase hereda los atributos de la superclase Personal_Universitario() y añade un nuevo atributo 'puesto'.
    El argumento 'puesto' es de tipo string y define el puesto de trabajo que desempeña esta persona."""
    # Constructor
    def __init__(self, id, nombre, email, puesto):
        super().__init__(id, nombre, email)
        self.__puesto = puesto
        print("Oficina creada.")

    # Getter y setter
    @property
    def puesto(self):
        return self.__puesto
    
    @puesto.setter
    def puesto(self, newValue):
        self.__puesto = newValue

    # Métodos
    def __str__(self):
        return super().__str__() + "\n\t-> Puesto de trabajo: {}.".format(self.puesto)


class Profesor(Personal_Universitario):
    """Subclase Profesor
    
    Esta subclase hereda los atributos de la superclase Personal_Universitario() y añade un nuevo atributo 'especialización'.
    El argumento 'especialización' es de tipo string y define la asignatura que imparte este profesor."""
    # Constructor
    def __init__(self, id, nombre, email, especializacion):
        super().__init__(id, nombre, email)
        self.__especializacion = especializacion
        print("Oficina creada.")

    # Getter y setter
    @property
    def especializacion(self):
        return self.__especializacion
    
    @especializacion.setter
    def pespecializacionuesto(self, newValue):
        self.__especializacion = newValue

    # Métodos
    def __str__(self):
        return super().__str__() + "\n\t-> Especialización del profesor: {}.".format(self.especializacion)
    

class Alumno(Personal_Universitario):
    """Subclase Alumno
    
    Esta subclase hereda los atributos de la superclase Personal_Universitario() y añade un nuevo atributo 'creditosAprobados'.
    El argumento 'creditosAprobados' es de tipo integer y representa los créditos aprobados hasta el momento por el alumno."""
    # Constructor
    def __init__(self, id, nombre, email, creditosAprobados):
        super().__init__(id, nombre, email)
        self.__creditosAprobados = creditosAprobados
        print("Oficina creada.")

    # Getter y setter
    @property
    def creditosAprobados(self):
        return self.__creditosAprobados
    
    @creditosAprobados.setter
    def creditosAprobados(self, newValue):
        self.__creditosAprobados = newValue

    # Métodos
    def __str__(self):
        return super().__str__() + "\n\t-> Creditos aprobados por el alumno: {}.".format(self.creditosAprobados)


# Creando algunas instancias de clase.
person1 = Personal_Universitario(234, "alvaro", "test@test.com")
oficinista1 = Oficina(345, "Nebraska", "nebri@test.com", "Oficinista")
profesor1 = Profesor(555, "Juana", "juana@test.com", "Historia")
alumno1 = Alumno(987, "Irene", "irene@test.com", 180)
# print(person1.datos) # Esto nos da error
#print(person1.__dict__)
# print(person1)
print(oficinista1)
print(profesor1)
print(alumno1)
print(alumno1.datos, alumno1.creditosAprobados)
print(alumno1.__dict__)
```

**3) Escoge la temática que prefieras y define una estructura de clases y subclases donde practiques la herencia en al menos dos niveles (herencia multi nivel).
```python
"""Para este ejercicio quiero plantear un sistema de herencia multinivel que vaya de una superclase Animal(), pasando por una subclase Pet() y extraer dos posibles subclases Dog() y Cat(), cada una recibe herencia múltiple de su clase padre Pet() y de Canine() y Feline() respectivamente."""

class Animal():
    """Clase Animal

    Argumentos:
    - Type: de tipo string. Define si el animal es mamífero, pez, ave...
    - Vertebrate: de tipo boolean. Define si tiene o no estructura osea.
    - Legs: de tipo integer. Define el numero de patas que tiene el animal.
    """
    # Constructor
    def __init__(self, type, vertebrate, legs):
        self.__type = type
        self.__vertebrate = vertebrate
        self.__legs = legs
        # print("Animal creado") # Comentamos este print para no repetir la estructura en el output, durante la contrucción del objeto.

    # Getters y setters
    @property
    def type(self):
        return self.__type
    @type.setter
    def type(self, newValue):
        self.__type = newValue

    @property
    def vertebrate(self):
        return self.__vertebrate
    @vertebrate.setter
    def vertebrate(self, newValue):
        self.__vertebrate = newValue

    @property
    def legs(self):
        return self.__legs
    @legs.setter
    def legs(self, newValue):
        self.__legs = newValue

    # Métodos
    def __str__(self):
        return """\n\nLos datos de este animal:
        -> Tipo: {}
        -> Vertebrado: {}
        -> Nº de patas: {}""".format(self.type, self.vertebrate, self.__legs)
   

class Pet(Animal):
    """Clase Pet
    
    Hereda los argumentos de la superclase Animal y añade argumentos nuevos
    Args:
    - Name: de tipo string. Damos un nombre a nuestra mascota
    - Vaccinate: de tipo boolean. Nuestra mascota puede o no estar vacunada.
    - Hair: de tipo string. Puede ser de pelo largo o corto, etc
    - Size: de tipo string. Mascota grande, mediana o pequeña
    - Gender: de tipo string. Puede ser macho, hembra o desconocido
    """
    # Constructor
    def __init__(self, type, vertebrate, legs, name, vaccinate, hair, size, gender):
        super().__init__(type, vertebrate, legs)
        self.__name = name
        self.__vaccinate = vaccinate
        self.__hair = hair
        self.__size = size
        self.__gender = gender
        #print("\n\nMascota creada")

    # Getters y Setters
    @property 
    def name(self):
        return self.__name
    @name.setter
    def name(self, newValue):
        self.__name = newValue

    @property 
    def vaccinate(self):
        return self.__vaccinate
    @vaccinate.setter
    def vaccinate(self, newValue):
        self.__vaccinate = newValue

    @property 
    def hair(self):
        return self.__hair
    @hair.setter
    def hair(self, newValue):
        self.__hair = newValue

    @property 
    def size(self):
        return self.__size
    @size.setter
    def size(self, newValue):
        self.__size = newValue

    @property 
    def gender(self):
        return self.__gender
    @gender.setter
    def gender(self, newValue):
        self.__gender = newValue

    # Métodos
    def isVaccinate(self):
        if self.vaccinate == True:
            return "Animal vacunado"
        else:
            return "Animal no vacunado"

    def __str__(self):
        return Animal.__str__(self) + "\n\nEsta mascota se llama: {}, su género es {}. Tiene el pelo {} y es {}. ¿Está vacunado? {}.".format(self.name,
                                                                                                                                                                 self.gender, 
                                                                                                                                                                 self.hair, 
                                                                                                                                                                 self.size, 
                                                                                                                                                                 self.isVaccinate())

# Las siguientes clases Canine y Feline no heredan de ninguna superclase
class Canine():
    """Clase Canine
    
    Emplearemos esta clase para definir los atributos clásicos que definen a los animales caninos como perros, lobos o coyotes.

    Args:
    - Smell: de tipo integer. Capacidad olfativa medida en relación a la distancia a la que puede oler a sus presas u objetivos. Por defecto 200m.
    - Hearing: de tipo integer. Capacidad auditiva medida en relación al número máximo de Hz que es capaz de percibir. Por defecto 40000Hz."""

    # Constructor
    def __init__(self, smell=200, hearing=40000):
        self.smell = smell
        self.hearing = hearing
    
    # Métodos
    def __str__(self):
        return "\nUn canino puede oler a sus presas hasta a {} metros. Desarrollan una capacidad auditiva de hasta {}Hz.".format(self.smell, self.hearing)

class Feline():
    """Clase Feline
    
    Emplearemos esta clase para definir los atributos clásicos que definen a los animales felinos como tigres, leones o gatos.

    Args:
    - NightVision: de tipo integer. Capacidad de visión en la oscuridad medida en metros. Por defecto será de 80m.
    - Claws: de tipo string. Son las garras del felino. Podrán ser afiladas o recortadas. Por defecto, estarán afiladas"""

    # Constructor
    def __init__(self, nightVision=80, claws="afiladas"):
        self.nightVision = 80
        self.claws = claws

    # Métodos
    def __str__(self):
        return "\nUn felino puede ver en la oscuridad hasta a {} metros de distancia. Este felino cuenta con garras: {}.".format(self.nightVision, self.claws)

class Dog(Canine, Pet):
    """Clase Dog
    
    Definimos la clase Dog. Sus objetos heredan los atributos de los caninos y de las mascotas.
    Args:
    - Race: de tipo string. Define el tipo de raza y morfología de este perro.
    - DogType: de tipo string. Define el tipo de perro que es: perro policia, perro bombero, perro domestico, perro concurso, perro callejero..."""
    # Constructor
    def __init__(self, type, vertebrate, legs, name, vaccinate, hair, size, gender, smell, hearing, race, dogType="doméstico"):
        Canine.__init__(self, smell, hearing)
        Pet.__init__(self, type, vertebrate, legs, name, vaccinate, hair, size, gender)
        self.__race = race
        self.__dogType = dogType

    # Getters y setters
    @property
    def race(self):
        return self.__race
    @race.setter
    def race(self, newvalue):
        self.__race = newvalue
    
    @property
    def dogType(self):
        return self.__dogType
    @dogType.setter
    def dogType(self, newvalue):
        self.__dogType = newvalue

    # Métodos
    def __str__(self):
        return Pet.__str__(self) + Canine.__str__(self) + "\nEste es un perro {} de raza {}. \n----------------------------".format(self.dogType, self.race)

class Cat(Feline, Pet):
    """Clase Cat
    
    Definimos la clase Cat. Sus objetos heredan los atributos de los felinos y de las mascotas.
    Args:
    - Raza: de tipo string. Define el tipo de raza y morfología de este gato
    - SleepHoursPerDay: de tipo integer. Define el número de horas que duerme el gato durante un día. Por defecto será de 14."""

    # Constructor
    def __init__(self, type, vertebrate, legs, name, vaccinate, hair, size, gender, nightVision, claws, race, sleepHoursPerDay=14):
        Feline.__init__(self, nightVision, claws)
        Pet.__init__(self, type, vertebrate, legs, name, vaccinate, hair, size, gender)
        self.__race = race
        self.__sleepHoursPerDay = sleepHoursPerDay

    # Getters y setters
    @property
    def race(self):
        return self.__race
    @race.setter
    def race(self, newvalue):
        self.__race = newvalue
    
    @property
    def sleepHoursPerDay(self):
        return self.__sleepHoursPerDay
    @sleepHoursPerDay.setter
    def sleepHoursPerDay(self, newvalue):
        self.__sleepHoursPerDay = newvalue

    # Métodos
    def __str__(self):
        return Pet.__str__(self) + Feline.__str__(self) + "\nEste gato es de raza {} y duerme una media de {} horas al día.\n----------------------------".format(self.race, self.sleepHoursPerDay)

#### Pruebas de creación de instancias de clase para Animal, Pet, Canine y Feline ####
# animal1 = Animal("mamifero", True, 4)
# print(animal1)

# pet1 = Pet("mamifero", True, 4, name="Nebraska", hair="largo", size="grande", gender="hembra", vaccinate=True)
# print(pet1.vaccinate)
# print(pet1.__dict__)
# print(pet1)

# canine1 = Canine()
# print(canine1)
# feline1 = Feline()
# print(feline1)

Nebraska = Cat(type="mamifero", vertebrate=True, legs=4, name="Nebraska", hair="largo", size="grande", gender="hembra", vaccinate=True, nightVision=80, claws="afiladas", race="Europea")
print(Nebraska)
print(Nebraska.__dict__)

Max = Dog("mamifero", True, 4, "Max", False, "largo", "mediano", "macho", smell=200, hearing=40000, race="pastor alemán")
print(Max)
print(Max.__dict__)

#### Vamos a crear dos métodos relacionados a la alimentación de perros y gatos y otro para las visitas al veterinariodogType

# El método alimentación consulta la clase y otros atributos del animal para definir cual puede ser la mejor opción de alimentación
def alimentoIdeal(my_pet, size):
    """Esta función toma como argumentos el tipo de mascota y el tamaño como argumentos para definir el mejor alimento"""
    #print(my_pet, size)
    pet_class = type(my_pet).__name__

    print("\n----------------------\nLa comida ideal para: {}\nTeniendo en cuenta que tu mascota es {} y su tipo:".format(my_pet.name, size))
    if pet_class == 'Dog' and size == 'pequeño':
        print("Lo mejor para un perro de este tamaño es pienso húmedo de la marca XXXXXX.")
    elif pet_class == 'Dog' and size == 'mediano':
        print("Te recomendamos las croquetas de pollo y verduras.")
    elif pet_class == 'Dog' and size == 'grande':
        print("Lo mejor para evitar que tu perro engorde es pienso dietético.")
    elif pet_class == 'Cat' and size == 'pequeño':
        print("Debes alimentarla con leche vaporizada y pienso para cachorros")
    elif pet_class == 'Cat' and size == 'grande':
        print("La mejor opción es pienso dietético combinado con algo de comida húmeda.")
    ## Consideramos que la mascota no sea ni de la clase Dog o Cat
    elif pet_class != 'Dog' or pet_class != 'Cat':
        print("Parece que no tenemos un dieta específica para este tipo de mascota.")
    else:
        print("Ha habido un error.")

# El método visitVet toma un argumento y comprueba si tiene las vacunas puestas.
def visitVet(my_pet):
    
    if my_pet.vaccinate == True:
        print("\nTodo está en orden con tu mascota {}.\n---------------------".format(my_pet.name))
    else:
        print("\nDebes vacunar a tu masota {} para mantener su buen estado de salud.\n---------------------".format(my_pet.name))
        user_choice = input("¿Quieres que vacunemos a {}? Y/n".format(my_pet.name))

        if user_choice == 'y' or user_choice == 'Y':
            return print("\tHemos vacunado a {}!".format(my_pet.name))
        else:
            return print("\tTe recomendamos hacerlo. Puedes consultar nuestros servicios en cualquier momento. Muchas gracias!")


NebraskaIsCat = type(Nebraska).__name__
MaxIsDog = type(Max).__name__

visitVet(Nebraska)
visitVet(Max)

alimentoIdeal(Nebraska, Nebraska.size)

# Prueba de acceso a size
size = Nebraska.size
print("\n--------\nTamaño de Nebraska:", size)



```
